# Understanding this Data

```python
df['Month of Order Date'].str[:4]

# we get basically an aggregation of the data by the year by only pulling out the 4th character of the column(from start)
```

```python
df.groupby(df['Month of Order Date'].str[:4]).describe()

# we will get a detailed dummary of the data by the year
```

```python
df.groupby(df['Month of Order Date'].str[:4]).describe().reset_index().rename(columns={'Month of Order Date':'Year'})

# we will now rename this column and add the indexing for it
```

```python
yearly_product_stats = df.groupby([df['Month of Order Date'].str[:4],"Product Name"]).sum().reset_index()
```

```python
yearly_product_stats.rename(columns={'Month of Order Date':'Year'})

# here we have a detailed report for the yearly product stats with the year and the product name
```

```python
df.groupby("Product Name").sum().reset_index()

# we get the total sales for each product
```

```python
yearly_product_stats.to_csv("YearlyProductStats.csv",header=True,index=False,encoding="utf-8")

# we save the data frame to a csv file
```

```python
yearly_product_stats.to_excel("YearlyProductStats.xlsx",header=True,index=False)

# we save the data frame to an excel file
```

# Visualizing the Data

```python
import matplotlib.pyplot as plt
```

```python
df.groupby(df["Month of Order Date"]).sum().plot()

# we plot the data frame
```

```python
df.groupby(df["Month of Order Date"]).sum().plot(kind="hist" ,bins=10)

# plot a histogram (DISTRIBUTION OF SALES BY MONTH)
```

```python
df.groupby(df["Product Name"]).sum().sort_values(by="Sales").tail(10).plot(kind="barh")

# we get the top 10 products with the highest sales
# sort_by is done in ascending order by default
```
